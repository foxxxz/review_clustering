### CounterVcetorizer

```python
distortions = []
for i in range(1, 21):
    print(i)
    km = KMeans(n_clusters=i,
                init='k-means++',
                max_iter=300,
                n_init=10,
                tol=0.0001)
    km.fit(v0)
    distortions.append(km.inertia_)
    
plt.plot(range(1,21), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

```python
cnum = 6
```

```python
km = KMeans(n_clusters=cnum,
            init='k-means++',
            max_iter=300,
            n_init=10,
            tol=0.0001)
km.fit(v0)

print("Silhouette Coefficient: %0.3f"
      % silhouette_score(v0, km.labels_))
```

```python
dfRes = pd.DataFrame(df1.index.values, columns=['uid'])
dfRes['label'] = km.labels_
for i in range(cnum):
    print('label{}: {}'.format(i, len(dfRes[dfRes.label == i])))
```

```python
term = np.array(vect0.get_feature_names())
t = []
km.cluster_centers_.argsort()[0]

for i in range(cnum):
    t.append(term[km.cluster_centers_.argsort()[i]].tolist())
    pprint(t[i][:10])
```

```python
def vans_tokenizer(corpus):
    pos = twitter.pos(corpus)
    res = [x[0] for x in pos if (x[1] == u'Noun' or x[1] == u'Verb' or x[1] == u'Adjective' or x[1] == u'Suffix')]
    return res
```

```python
pprint(vans_tokenizer(corpus[4]))
```

```python
vect = CountVectorizer(tokenizer=vans_tokenizer,
                         analyzer='word',
                         stop_words=mask)
```

```python
%%time
vect.fit(corpus)
```

```python
v = vect.transform(corpus).toarray()
```

```python
cnum = 8
km = KMeans(n_clusters=cnum,
            init='k-means++',
            max_iter=300,
            n_init=10,
            tol=0.0001)
km.fit(v)

print("Silhouette Coefficient: %0.3f"
      % silhouette_score(v, km.labels_))
```

```python
dfRes = pd.DataFrame(df1.index.values, columns=['uid'])
dfRes['label'] = km.labels_
for i in range(cnum):
    print('label{}: {}'.format(i, len(dfRes[dfRes.label == i])))
```

```python
term = np.array(vect.get_feature_names())
t = []
km.cluster_centers_.argsort()[0]

for i in range(cnum):
    t.append(term[km.cluster_centers_.argsort()[i]].tolist())
    pprint(t[i][:10])
```

```python
vect00 = CountVectorizer(tokenizer=twitter.nouns,
                         analyzer='word',
                         stop_words=mask)

vect11 = TfidfVectorizer(tokenizer=twitter.nouns,
                         analyzer='word',
                         stop_words=mask)
```

```python
%%time
vect00.fit(corpus)
vect11.fit(corpus)
```

```python
v00 = vect00.transform(corpus).toarray()
v11 = vect11.transform(corpus).toarray()
```

```python
distortions = []
for i in range(1, 21):
    km0 = KMeans(n_clusters=i,
                init='k-means++',
                max_iter=300,
                n_init=10,
                tol=0.0001)
    km0.fit(v00)
    distortions.append(km0.inertia_)
    
plt.plot(range(1,21), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```
