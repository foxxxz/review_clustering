# 형태소 분석

```python
twitter = Twitter()
```

```python
def nav_tokenizer(corpus):
    pos = twitter.pos(corpus)
    res = [x[0] for x in pos if (x[1] == u'Noun' or x[1] == u'Adjective' or x[1] == u'Verb')]
    return res
```

```python
stopwords = [u'피', u'지오', u'겔', u'좋', u'짱', u'굿', u'너무', u'쓰', u'역시']
```

```python
vect = TfidfVectorizer(tokenizer=nav_tokenizer, stop_words=stopwords)
vect.fit(corpus)
print("token n/a/v: %d" % len(vect.vocabulary_))
```

```python
X = vect.transform(corpus).toarray()
```

# Doc2Vec

```python
from gensim.models.doc2vec import TaggedDocument
```

```python
train_docs = [(nav_tokenizer(x), 0) for x in corpus]
```

```python
pprint(train_docs[0][0])
```

```python
tagged_train_docs = [TaggedDocument(d, [c]) for d, c in train_docs]
```

```python
from gensim.models import doc2vec
```

```python
doc_vectorizer = doc2vec.Doc2Vec(size=300, alpha=0.025, min_alpha=0.025, seed=1234)
doc_vectorizer.build_vocab(tagged_train_docs)
doc_vectorizer.train_lbls = False
doc_vectorizer.train_words = True
```

```python
%time
for epoch in range(10):
    doc_vectorizer.train(tagged_train_docs, total_examples=len(train_docs), epochs=epoch)
    doc_vectorizer.alpha -= 0.002 # decrease the learning rate
    doc_vectorizer.min_alpha = doc_vectorizer.alpha # fix the learning rate, no decay
```

```python
pprint(doc_vectorizer.most_similar(u'가족'))
```

```python
pprint(doc_vectorizer.most_similar(positive=[u'가족', u'아이'], negative=[u'남편']))
```

```python
train_x = [doc_vectorizer.infer_vector(doc.words) for doc in tagged_train_docs]
```

* k-means w/ doc2vec

```python
km = KMeans(n_clusters=10, init='k-means++', max_iter=150, n_init=10, tol=0.0001)
km.fit(train_x)
print("Silhouette Coefficient: %0.3f" % silhouette_score(train_x, km.labels_))
```

```python
dfRes = pd.DataFrame(corpus, columns=['text'])
dfRes['label'] = km.labels_

for i in range(10):
    print('label{}: {}'.format(i, len(dfRes[dfRes.label == i])))
```

# LDA(Latent Dirichlet allocation)

```python
def remove_stopwords(words, stopwords):
    res = [x for x in words if x not in stopwords]
    return res
```

```python
text_ps = [nav_tokenizer(x) for x in corpus0]
text_ps = [remove_stopwords(x, stopwords) for x in text_ps]
dict_ps = corpora.Dictionary(text_ps)
dict_ps.save('physiogel.dict')  # save dictionary to file for future use
```

```python
len(dict_ps)
```

```python
#calulate TF-IDF
tf_ps = [dict_ps.doc2bow(x) for x in text_ps]
tf_model = models.TfidfModel(tf_ps)
tfidf_ps = tf_model[tf_ps]
corpora.MmCorpus.serialize('ps.mm', tfidf_ps)
```

```python
# train topic model
ntopics, nwords = 8, 5
```

```python
# LSI
lsi = models.lsimodel.LsiModel(tfidf_ps, id2word=dict_ps, num_topics=ntopics)
pprint(lsi.print_topics(num_topics=ntopics, num_words=nwords))
```

```python
# LDA
np.random.seed(0)  # optional
lda = models.ldamodel.LdaModel(tfidf_ps, id2word=dict_ps, num_topics=ntopics)
pprint(lda.print_topics(num_topics=ntopics, num_words=nwords))
```

```python
#HDP
np.random.seed(0)  # optional
hdp = models.hdpmodel.HdpModel(tfidf_ps, id2word=dict_ps)
pprint(hdp.print_topics(num_topics=ntopics, num_words=nwords))
```
